# **Introduction to Python Classes**

Previously, you have learned how to use **variables** to store data and **functions** to perform operations.

* **Variables** hold information (strings, integers, lists, tuple, dictionaries).
* **Functions** perform asctions - like taking inputs (parameters), performing tasks, and returning outputs.

**The Problem with Separating Data and Functions**
* This works for simple scripts. However, this approach becomes difficult to manage when programs grow larger, especially when building real-world systems like web apps or data pipelines.

* In real-world systems, we deal with complex "things" that have both **data** and **behavior** inextricably linked.

Let's explore the problem with a real-world case, and then get to the solutions.

## A Real-World Scenario: A Dog Daycare & Boarding Web Application

### **1. The "Dictionary" Approach (Functional but Messy):**

Imagine we are building a backend for a **Dog Daycare System**. We need to track **Users**, their **Dogs**, and their **Bookings**.

**High-Level Workflow**
1.  A user registers on the website.
2.  The system creates a `User` profile with their details like name, email, phone number, etc.
3.  The user adds one or more dogs to their profile.
4.  The user books daycare or boarding for a dog.
5.  The system stores and manages these relationships.

In a web application, our data is much more complex than a few strings.

* Users need **unique identifiers** (like a database ID).
* Users own dogs, meaning we need **nested data structures** (lists of dictionaries inside our main dictionary).
* Users take actions, like making **bookings**, which need to be validated against allowed services.

Let's build this out using only our "dictionary and function" approach.

==============================================================================================

#### **1.1. Setup and Logic**

First, we need to define the "rules" of our system. We will import a module to create unique IDs for each user and write a function that handles the logic for bookings. Notice that the functions are "floating"—it isn't attached to any specific user.

````python
from uuid import uuid4  # This generates a "Universally Unique Identifier" (a long, random string)

# 1. Global Configuration: The valid services our daycare offers
# Make a list of the valid services and use it for validation in the add_booking function below.
# This list exists outside of users.
# The program has to "look up" these values every time it wants to check if a booking is valid.
SERVICE_TYPES = ["daycare", "boarding"]

# 2. The Behavior: A standalone function to handle bookings
# This function is a "contractor." It doesn't belong to any user; Logic is floating separately from the data.
# It just operates on the Data (dictionaries), taking a dictionary and modifying it.
# This function requires a user dictionary, a valid service type, and a list of dates in list form to work. 
# If we forget to pass the correct arguments, it won't work!

def add_booking(user_dict, service_type, dates):
    # Step A: Validation
    # We check the global list to see if the requested service is allowed.
    if service_type not in SERVICE_TYPES:
        print(f"Error: '{service_type}' is not a valid service.")
        return
    
    # Step B: Updating the Data
    # We loop through the list of dates provided. 
    # For each date, we create a new booking string and append it to the user's bookings list.
    for date in dates:
        new_booking = f"{service_type} on {date}"
        user_dict["bookings"].append(new_booking)
        print(f"Success: Added {new_booking} for {user_dict['name']}")

# Another function to update the phone number. 
# This function also operates on the user dictionary, but it is separate from the add_booking function.
def update_phone(user_dict, new_number):
    user_dict["phone"] = new_number
    print(f"Updated {user_dict['name']}'s phone number to {new_number}.")
````

#### **1.2. Creating Our First User (The Data)**

Now, let's create our first user using a standard Python Dictionary.

In this stage, we are acting as a "Manual Data Entry" clerk. We have to be extremely careful. If we make a single typo in a key name (like typing "user_id" instead of "id" or "booking" instead of "bookings"), our future functions might not be able to find the information they need.

Notice how we have to manually call `uuid4()` and manually initialize empty lists bookings.

````python
# 3. Adding user 01 Data (The "Jacqui" Object represented as a Dictionary)
# We are manually defining the "State" (the data) of this user as a dictionary, by defining every single key and value.
user_01 = {
    "id": str(uuid4()),  # We call the uuid function to generate a unique ID for this user. If we forget this line, we won't have an ID!
    "name": "Jacqui",
    "email": "jacqui@example.com",
    "phone": "555-0100",
    # the value of "dogs" is a list of dictionaries - each dictionary represents one dog as one user can have more than one dog.
    # This is an example of "Nested Data" dictionary within dictionary... Hope we don't make a typo in the keys!
    "dogs": [
        {"name": "Waffles", "breed": "Corgi", "age": 3, "weight": 28}
    ],
    # "bookings" is initialized as an empty list. 
    # If we forget this line, the 'add_booking' function will CRASH later 
    # because it won't find a list to append to!
    "bookings": []
}

# Let's verify our data entry was successful by printing the dictionary
print("--- User 01 Profile Created ---")
print(f"User: {user_01['name']}")
print(f"ID assigned: {user_01['id']}")
print(f"Initial Bookings: {user_01['bookings']}")
print(f"\nUser Data: {user_01}")
````

#### **1.3. Executing an Action (Connecting Logic to Data)**

Now that we have the **Logic** (the `add_booking` function from Section 1) and the **Data** (the `user_01` dictionary from Section 2), we need to make them work together.

Because they are separate, we must "inject" the data into the function. We do this by passing `user_01` as an argument. The function then reaches inside the dictionary, finds the `"bookings"` list, and modifies it.

````python
# Usage 1: Adding bookings for user 01 Jacqui (Single and Multi-day)
# We have to "plug" the 'user_01' dictionary INTO the add_booking function. 
# This tells the function which user's bookings to update!
print(f"--- Processing Bookings for {user_01['name']} ---")

# We pass: (1) The dictionary, (2) The service string, (3) A list of date strings
add_booking(user_01, "daycare", ["2025-10-12"])
add_booking(user_01, "boarding", ["2025-11-20", "2025-11-21", "2025-11-22"])

print("-" * 30)
# Let's check the 'bookings' key inside user_01 to see if the function worked
print(f"Verification: {user_01['name']} now has {len(user_01['bookings'])} items in their booking history.")
print(f"Bookings List: {user_01['bookings']}")


# Usage 2: Updating Jacqui's phone number using the update_phone function
update_phone(user_01, "555-9999")
print("\n" + "-" * 30)
print(f"Updated Phone Number for {user_01['name']}: {user_01['phone']}")
````

#### **1.5. Conclusion: Architectural Limitations of the "Dictionary and Function" Approach**

While we have successfully built a working prototype using dictionaries and standalone functions, this approach presents significant challenges for **System Design** and scalability. Dictionaries are efficient for simple key-value storage, but they lack the formal structure required for robust applications.

### Key Architectural Risks:

1.  **Lack of Schema Enforcement**: Python dictionaries do not enforce a consistent schema. If `user_01` contains a `"phone"` key but `user_02` does not, the application becomes prone to `KeyError` exceptions later in the data pipeline.
2.  **Manual Redundancy and Maintenance**: Creating data structures manually is inefficient. If a requirement changes—for example, renaming the `"phone"` key to `"phone_number"`—every dictionary instance and every function referencing that key must be updated manually, increasing the risk of regression errors.
3.  **Data Inconsistency**: Loose dictionaries allow for subtle errors, such as using the plural `"dogs"` in one instance and the singular `"dog"` in another. This inconsistency leads to fragile retrieval logic that may fail during runtime.
4.  **Decoupled Logic and State**: In this functional approach, the logic (the function) and the state (the dictionary) are entirely separate. The function has no intrinsic connection to the data; it only operates on whatever object is explicitly passed to it.
5.  **High Cognitive Load**: Developers must maintain a "mental map" of the entire nested data structure. To write even a single line of code, you must remember the exact hierarchy, data types, and key names of the objects involved.

### The Complexity of Data Retrieval

Constructing the data is only the first hurdle; **consuming** that data is often more difficult. Imagine a requirement to generate a summary of a user's first dog and their most recent booking. To implement this, a developer must navigate multiple levels of nesting:

* Access the `"dogs"` list and verify it is not empty.
* Access the first element of that list and retrieve the `"name"` key.
* Access the `"bookings"` list and retrieve the last element (index `-1`).

Without a formal structure, the developer must constantly check for the existence of keys and the validity of indices to prevent the program from crashing.

## **2. The Solution: Python Classes and Object-Oriented Design**

### **2.1. Introduction**

To resolve the architectural issues of loose data structures, we require a mechanism to **enforce a consistent schema**. We must bundle state (data) and behavior (logic) into a single, cohesive unit.

In Python, this is achieved through **Classes**. A Class serves as a formal blueprint that guarantees every object is initialized with the correct attributes, automates repetitive tasks (like ID generation), and encapsulates logic through **Methods**.

This approach provides two primary advantages:
* **Schema Consistency**: It ensures every user instance possesses the same attributes (e.g., name, email, dog list).
* **Syntactic Clarity**: It allows us to access data via **Dot Notation** (e.g., `.name`) instead of string-based keys and brackets (e.g., `["name"]`), which reduces typos and improves code readability.

#### **Conceptual Shift: From Functional to Object-Oriented**

**Table 01: Structural Evolution**

| Concept | Functional Programming | Object-Oriented Programming (OOP) |
|---------|-----------------------|---------------------|
| **Data Storage** | Independent Variables / Dictionaries | **Attributes** (Variables attached to an object) |
| **Logic/Action** | Independent Functions | **Methods** (Functions attached to an object) |
| **Organization** | Decoupled code blocks | **Encapsulation** (A single, structured unit) |

**Table 02: Core Terminology**

| Concept  | Python Term | Definition |
| -------- | ----------- | ------------------------------------------------------------------- |
| **Blueprint** | `class` | The definition of a data type (e.g., the concept of a "User"). |
| **Instance** | Object | A specific realization of a class (e.g., the specific user "Jacqui"). |
| **State** | Attribute | A variable that resides within an object (e.g., `jacqui.email`). |
| **Behavior** | Method | A function that belongs to an object (e.g., `jacqui.update_phone()`). |

....
> **Core Principle**
> * **Class:** Defines the structure and capabilities of a data type.
> * **Object:** A concrete instance of that class with unique data values.

### **2.2. Class Design Best Practices**

Before we write the code for our first class, we must establish the industry standards for Python object-oriented design.

#### **Naming Conventions (PEP 8)**
Python utilizes strict naming conventions to help developers instantly distinguish between structural blueprints and runtime data:
* **Classes use `PascalCase`**: Every word begins with a capital letter, with no underscores (e.g., `User`, `DogProfile`, `DaycareSystem`).
* **Variables, Attributes, and Methods use `snake_case`**: All lowercase letters with underscores separating words (e.g., `user_01`, `user_id`, `update_phone()`). 

#### **Responsibility: Where should the logic live?**
In functional programming, logic floats independently. In Object-Oriented Programming, we must decide which class *owns* a specific behavior. This is governed by the **Single Responsibility Principle**.
* **State Modification**: A method should reside within the class whose data it primarily modifies. For example, `update_phone()` mutates a user's contact information, so it inherently belongs in the `User` class.
* **Internal Calculation**: If a system needs to categorize a dog's size based on its weight, the `Dog` class should own a `get_size_category()` method. 
* **Interaction**: When a User books a Daycare session for a Dog, we must carefully decide where `process_booking()` lives. Initially, we will place it in `User` to track account history.

#### **Determining the Implementation Order: Entity Dependencies**

Which class should we architect first: `User` or `Dog`?

In system design, we answer this by analyzing **Dependencies**. 
* A **User** can register for our platform without currently registering a dog (for instance, they may simply want to create an account to view services first). Their `dogs` list can safely initialize as empty.
* Conversely, a **Dog** cannot exist in our daycare system as an orphan. For liability and billing, a dog record absolutely requires an `owner_id` to link it to a human.

Because the `Dog` entity is dependent on the prior existence of a `User` entity, **we must define and implement the `User` class first.**

### **2.3. Implementing Classes**

The following implementation shows the standard architecture of a Python class. This "Manual" definition allows us to control the **Initialization** process, ensuring that every object enters the system in a valid state.

#### **Architectural Improvements over the Dictionary Approach:**

*  **Automated State Initialization**: The `user_id` is generated automatically during instantiation. This removes the risk of a developer forgetting to assign a unique identifier or manually duplicating one.
*  **Schema Enforcement**: By requiring specific parameters during creatio|n, we ensure that every user consistently possesses a name, an email, and a designated list of pets.
*  **Encapsulation of Logic**: The `process_booking` logic is no longer a "floating" function. It is now a **Method** defined within the class. This allows the object to operate on its own internal state (`self`) and perform validation—such as verifying the dog's name exists—before modifying the booking history.

#### **Defining Attributes and Methods**

To build this class, we define **Attributes** (the internal data storage):
* **Dynamic Attributes**: Values provided via parameters during initialization, such as `name`, `email`, and `phone`.
* **Automatic Attributes**: Values generated or initialized internally, such as the `user_id` and the `bookings` list.

#### **Functional Logic within Methods**
In the implementation below, note that the `process_booking` method is designed for batch processing. It accepts a `list` of dates and utilizes a `for` loop to append each service entry to the object's internal booking history. This encapsulates the "How" of the booking process within the object itself.

==============================================================================================

**Next Step:** Below is the Python implementation defining our primary system entities.

````python
from uuid import uuid4

SERVICE_TYPES = ["daycare", "boarding"]

class User:
    """
    Represents a human user in our application.
    This class encapsulates identity, contact details, and booking history.
    """
    
    def __init__(self, name: str, email: str, dogs_names: list | None = None, phone: str | None = None):
        """
        The 'Constructor': Sets up the initial state of the object.
        
        Parameters
        ----------
        name : str
            The user's full name.
        email : str
            The user's email address.
        dogs_names : list
            A list of names of the dogs owned by this user (e.g., ["Millie"] or ["Waffles", "Tillie"]).
        phone : str | None, optional
            The user's phone number. Defaults to None.
        """
        # 1. DYNAMIC ATTRIBUTES (State provided at creation)
        self.name = name
        self.email = email
        self.phone = phone
        
        # 2. DATA STRUCTURES (Initialized automatically)
        # Ensures the instance has a reliable list structure for dogs and bookings
        # We handle the 'None' case to ensure self.dogs is always a list
        self.dogs = dogs_names if dogs_names is not None else [] # Will hold a list of registered dog names (or Dog objects)
        self.bookings = []  # Initializes as an empty list to prevent AttributeErrors
        
        # 3. AUTOMATED METADATA
        # Automatically assigns a unique identifier upon instantiation
        self.user_id = str(uuid4())

    def update_phone(self, new_number: str):
        """
        Updates the phone attribute for the current instance.
        """
        self.phone = new_number
        print(f"--- [System Update] ---")
        print(f"Updated {self.name}'s contact number to: {self.phone}")
        
    def process_booking(self, dog_name: str, service_type: str, date_list: list):
        """
        Records a booking for a specific dog on one or more dates.
        Encapsulates validation and state modification logic.
        """
        print(f"--- Attempting Booking: {service_type} for {dog_name} ---")
        
        # 1. VALIDATION: Verifies dog ownership within the instance's own data
        if dog_name not in self.dogs:
            print(f"❌ Error: {self.name} does not have a dog named '{dog_name}'. Booking failed. Please register the dog first.")
            return

        # 2. PROCESSING: Iterates through the provided date list
        for date in date_list:
            # Constructs a structured string record
            entry = f"{date} | {dog_name} | {service_type}"
            
            # Updates the instance's internal booking history
            self.bookings.append(entry)
            print(f" ✅ Confirmed for {date}")


"""
Now let's define the `Dog` class. 

**Key Design Choice 1: The Foreign Key**
Notice we are adding a parameter called `owner_id`. 
* This creates a link between the **Dog** and the **User**.
* In database design, this is called a **Foreign Key**. It tells us exactly who this dog belongs to.

**Key Design Choice 2: Internal Logic (Methods)**
Let's have a logic to determine if a dog was "Small", "Medium", or "Large" based on weight.
Instead of writing that `if-else` logic every time we need it, we can put it **inside** the class as a method called `get_size_category`.

"""

class Dog:
    """
    Represents a specific dog profile in the system.
    Must be linked to a User via owner_id.
    """

    def __init__(self, name: str, breed: str, weight: float, age: float, owner_id: str):
        # 1. DYNAMIC ATTRIBUTES
        self.name = name
        self.breed = breed
        self.weight = weight
        self.age = age
        self.owner_id = owner_id  # Relational link to User.user_id
        
        # 2. AUTOMATED METADATA
        self.dog_id = str(uuid4()) 
        
    def get_size_category(self):
        """
        Calculates and returns the size category based on the instance's weight attribute.
        """
        if self.weight < 15:
            return "Small"
        elif 15 <= self.weight < 40:
            return "Medium"
        else:
            return "Large"
````

### **2.4. Deconstructing the Class Anatomy**

Looking at the code above, several new Python keywords and concepts were introduced. Let's break down the mechanical components of a class to understand how it transforms data into an object.

Let's break down the mechanical components of the Python classes we just built to understand how they transform raw data into structured objects.

#### **External Modules (`uuid`)**
```python
from uuid import uuid4
```
A module is a file containing pre-written Python code (variables, functions, classes) that we can bring into our project. Here, we import the `uuid4` function from the uuid (Universally Unique Identifier) module to automatically generate secure, unique ID strings for our users and dogs.

#### **The Constructor: `__init__`**
```python
def __init__(self, name: str, email: str, phone: int | None = None):
```

The `__init__` method is a special built-in Python function known as the Constructor (or initializer).

* Note the double underscores (__) on both sides. In Python, these are colloquially called "dunder" methods, signaling that they are reserved by the system for background behaviors.

* You rarely call __init__ directly. Instead, Python triggers it automatically the moment you instantiate a new object.

* Its primary responsibility is to establish the initial **State** of the object (e.g., creating empty lists so the object is ready to be used safely).

#### **The Instance Reference: `self`**
The word `self` is a fundamental concept in Object-Oriented Programming.

* `self` acts as a placeholder for the specific instance currently being manipulated.

* Think of the Class as a blank template, and `self` as the specific document currently being filled out.

* When you create `user_01 = User("Jacqui", ...)`, Python passes the `user_01` object into the `__init__` method as the self argument. This ensures that Jacqui's data is written to Jacqui's memory space.

#### **Type Hints and Default Values**
In the parameter list `(name: str, email: str, phone: str | None = None)`, we use modern Python features to make our code robust:

* Type Hints (`: str`): These indicate the expected data type. It documents that name and email must be strings.

* Optional Parameters (`str | None`): This syntax means the value can be a string OR it can be None.

* Default Values (= None): If a developer creating a User does not provide a phone number, Python automatically assigns `None` (representing an intentional absence of value) rather than crashing the program.

### Parameters vs. Attributes (A Critical Distinction)

When working with classes, it is very important to distinguish between **parameters** and **attributes**.  
Although they often share identical names in the `__init__` method, they play entirely different roles in system memory.

**Parameters (The Temporary Inputs)** Parameters are temporary variables defined in the function signature: def `__init__(self, name, email):`

* They act as temporary carriers for the input data (e.g., passing "Jacqui" into `name`).
* They exist only while the `__init__` function is executing. Once the setup is complete, these variables vanish from memory.

**Attributes (The Permanent Storage):** Attributes are the variables attached to the object using the `self` keyword: `self.name`

* They are the permanent memory of the object (`user_01.name`, `user_01.email`).

* They persist for the entire lifetime of the object and can be accessed or modified later using dot notation.

**How Parameters Become Attributes**

Inside `__init__`, we explicitly map the temporary inputs to the permanent storage:

```python
self.name = name
```

This single line of code is the bridge between the two concepts. It translates to:
> "Take the value currently held in the temporary parameter `name`, and store it permanently in this specific object's attribute called `self.name`."
|
> **Core Rule: Parameters are temporary inputs to a function; Attributes are the persistent state of an object.**

### **2.5. Using the User Class: Instantiation and Dot Notation**

To utilize the `User` class blueprint, we must instantiate concrete **Objects** (instances). This process is known as **Instantiation**. 

Instead of manually constructing a dictionary, we invoke the class by its name and pass in the required initialization arguments. The Python interpreter subsequently executes the `__init__` constructor and allocates a distinct object in memory.

#### **Execution Flow and System Simplification**
When an object is instantiated and its methods are invoked, the architectural benefits become immediately apparent:

* **Internal State Management**: The `self` parameter dynamically binds to the specific instance being operated on. Validation logic—such as verifying dog ownership within `process_booking`—operates exclusively on that instance's internal data, preventing cross-contamination between users.
* **Schema Consistency**: Manual data structure replication is eliminated. Invoking the class guarantees that every new user adheres strictly to the predefined schema.
* **Custom Data Typing**: The `User` class now functions as a natively recognized data type within the application, operating with the same predictability as built-in types like `int` or `list`.
* **Standardized Interfaces**: Methods like `process_booking` strictly define their input parameters (e.g., requiring a `list` for dates). This ensures internal loops and logic function uniformly across all instances.

#### **Key Technical Shift: Interaction via Dot Notation**

Instead of passing loose data structures into detached functions, actions are triggered directly on the object. This exemplifies **Encapsulation**: the data and the logic are formally bound together.

* In the prior "Dictionary" approach, data was accessed via string-based key lookups: `user_01["name"]`.
* In the "Class" approach, data is accessed via **Dot Notation**: `user_01.name`.

**Dot Notation** provides critical system design benefits:
*  **Syntactic Clarity**: A call like `user_01.process_booking()` explicitly denotes the entity performing the action.
*  **Self-Sufficiency**: The object inherently retains the state required to execute its methods; external data injection is minimized.
*  **Namespace Isolation**: The `name` attribute of `user_01` occupies a completely separate memory address from the `name` attribute of `user_02`.

#### **Positional vs. Keyword Arguments in Instantiation**

When instantiating an object, Python permits arguments to be passed positionally. For example, you could write:

```python
user_01 = User("Jacqui", "jacqui@example.com", ["Waffles"])
```

This relies entirely on the order of the parameters defined in `__init__`. While functionally valid, professional software development heavily favors **Keyword Arguments**:

```python
user_01 = User(name="Jacqui", email="jacqui@example.com", dogs_names=["Waffles"])
```

**Why we use Keyword Arguments in production environments:**
* **Explicit Data Mapping**: It prevents the accidental transposition of variables of the same data type. If an object requires multiple string inputs, positional arguments make it easy to accidentally pass an email into the name parameter without triggering a runtime error. Keyword arguments make this impossible.

* **Code Maintainability**: The instantiation code remains completely unambiguous. Future developers reviewing the code do not need to hunt down the original class constructor signature to understand what each string represents.


**Next Step:** In the following Python cell, we will instantiate our users using keyword arguments and execute the methods we have defined.

##### **Create Instances**

````python
# 1. INSTANTIATION: Creating the objects (instances)
# We pass the data into the blueprint. The __init__ method handles the rest.
user_01 = User(
    name="Jacqui", 
    email="jacqui@example.com", 
    phone="555-0100"
)
# Manually appending dog names as strings for this iteration
user_01.dogs = ["Waffles", "Remy"]

user_02 = User(
    name="Sarah", 
    email="sarah@example.com"
)
user_02.dogs = ["Millie"]

# 2. STATE VERIFICATION: Accessing data via Dot Notation
print(f"Name: {user_01.name}")
print(f"Email: {user_01.email}")
print(f"System ID: {user_01.user_id}")
````

#### **Executing Methods: Triggering Encapsulated Behavior**

Now we will utilize **Dot Notation** to trigger the methods we defined in our class. This demonstrates how an object manages its own internal state and enforces its own validation rules without relying on external, floating functions.

````python
print(f"--- Initial State for {user_01.name} ---")
print(f"Current Phone: {user_01.phone}")

# 1. STATE MODIFICATION: Updating an attribute via a method
user_01.update_phone("555-9999")

print("\n--- Processing Bookings ---")

# 2. VALID EXECUTION: Processing a legitimate booking
# The object 'user_01' verifies against its own 'dogs' list. "Waffles" is present.
user_01.process_booking(
    dog_name="Waffles", 
    service_type="daycare", 
    date_list=["2026-03-01", "2026-03-02"]
)

# 3. ENCAPSULATED VALIDATION: Attempting an invalid booking
# "Millie" belongs to user_02. user_01's internal validation will intercept and block this.
user_01.process_booking(
    dog_name="Millie", 
    service_type="boarding", 
    date_list=["2026-04-10"]
)

# 4. RETRIEVAL: Verifying the internal booking history array
print(f"\n--- Final Booking History for {user_01.name} ---")
for record in user_01.bookings:
    print(f" - {record}")
````

### **2.7. Dynamic Lifecycle: Updating Objects Post-Instantiation**

In a real-world application, data is rarely static. A user might sign up today but not register their pet until next week. Our updated `User` class architecture now supports this **Dynamic Lifecycle** by allowing the `dogs_names` list to be optional during initialization.

#### **Handling Optional Data**
By setting the default value to `None` (`dogs_names: list | None = None`), we prevent the system from crashing if a user is created without pets. Inside the constructor, we use a conditional expression:
`self.dogs = dogs_names if dogs_names is not None else []`

This ensures that even if no names are provided, the `self.dogs` attribute is initialized as an **empty list** rather than `None`. This is a critical safety measure; it allows us to call methods like `.append()` or check membership (`if dog in self.dogs`) later without triggering an `AttributeError`.

#### **The Workflow**
1. **Account Creation**: The `User` instance is created with only basic contact info.
2. **State Mutation**: As the user adds pets, we update the internal `self.dogs` list.
3. **Behavioral Activation**: Once the state is updated, the `process_booking` method (which previously would have failed) becomes functional for that specific pet.

````python
# 1. INITIAL SIGN-UP: User creates an account without a dog
user_03 = User(
    name="Marcus",
    email="marcus@example.com"
)

print(f"User Created: {user_03.name}")
print(f"Initial Dog List: {user_03.dogs}") # Displays as []

# 2. ATTEMPTING BOOKING (Expected Failure)
# Since the dogs list is empty, our validation logic should catch this.
user_03.process_booking("Buddy", "daycare", ["2026-07-01"])

# 3. REGISTERING A PET: Updating the instance state
# Marcus gets a dog and registers the name in the system
new_dog_name = "Buddy"
user_03.dogs.append(new_dog_name)

print(f"\n[Update] {user_03.name} registered a new pet: {new_dog_name}")
print(f"Updated Dog List: {user_03.dogs}")

# 4. RE-ATTEMPTING BOOKING (Expected Success)
# Now that the internal state contains "Buddy", the validation passes.
user_03.process_booking("Buddy", "daycare", ["2026-07-01"])

# 5. CREATING THE DOG OBJECT: Linking to the User
# We now create the formal Dog profile using Marcus's ID
buddy_profile = Dog(
    name="Buddy",
    breed="Beagle",
    weight=22.0,
    age=1.5,
    owner_id=user_03.user_id
)

print(f"\nDog Profile Created: {buddy_profile.name} (Owner ID: {buddy_profile.owner_id})")
````

### **2.8. Architectural Refactoring: The Booking Class**

As our system grows, we must continuously evaluate our design. Previously, we managed reservations by appending simple text strings (e.g., `"2026-03-01 | Waffles | daycare"`) into the `User.bookings` list. 

While functional for a basic script, this approach limits our system. A string cannot easily calculate revenue, update its own cancellation status, or link directly to a specific pet's medical records. 

To resolve this, we will introduce a dedicated **`Booking` Class**. This transforms a simple text label into a rich, structured **Transactional Object**.

#### **Defining the Transactional Entity**
The `Booking` class acts as the central hub that connects our existing entities. It requires several critical data points upon instantiation:
* `user_id`: The Foreign Key linking to the human responsible for billing.
* `dog_id`: The Foreign Key linking to the specific pet receiving the service.
* `service_type`: The category of service (e.g., daycare, boarding).
* `date`: The scheduled date of the reservation.

> **Industry Standard Note on Dates:** > In a professional production environment, dates are never stored as plain text strings. We use Python's built-in `datetime` module. Storing dates as proper objects allows the system to perform date math (e.g., `booking_date - today` to trigger a 48-hour reminder email) or chronological filtering. For the scope of this foundational lesson, we will use strings to maintain focus on Class mechanics.

#### **Encapsulated Business Logic**
Notice how the `Booking` class automatically derives the `price` attribute during instantiation. By passing in the `service_type`, the object references system constants to calculate its own cost, ensuring pricing remains consistent and protected from manual entry errors.

````python
from uuid import uuid4

# SYSTEM CONSTANTS (PEP 8 standard: All caps for constant values)
BOARDING_PRICE = 70.0
DAYCARE_PRICE = 50.0

class Booking:
    """
    Represents a single transactional reservation in the system.
    Connects a User, a Dog, and a specific Service Date.
    """
    
    def __init__(self, user_id: str, dog_id: str, service_type: str, date: str):
        # 1. PARAMETERS (Provided State)
        self.user_id = user_id
        self.dog_id = dog_id
        self.service_type = service_type.lower() # Standardize input to lowercase
        self.date = date
        
        # 2. INTERNAL BUSINESS LOGIC (Derived State)
        # The object automatically determines its price based on the service requested
        if self.service_type == "boarding":
            self.price = BOARDING_PRICE
        elif self.service_type == "daycare":
            self.price = DAYCARE_PRICE
        else:
            self.price = 0.0  # Fallback for unrecognized services
            print(f"[Warning] Unrecognized service type: {self.service_type}")
        
        # 3. AUTOMATED METADATA
        self.booking_id = str(uuid4())
        self.status = "Confirmed"  # Default state upon creation
````

### **2.8. Implementing Object Composition: Linking Bookings to Users**

Now that we have our `Booking` blueprint, we can observe **Object Composition** in action. Object Composition is a core design principle where complex objects (like our `User`) are built by combining simpler, specialized objects (like `Booking` and `Dog`).

Instead of storing a flat string like `"2026-03-01 | Waffles | daycare"` in the user's `bookings` list, we will store the entire `Booking` object. 



#### **The Advantage of Storing Objects**
When we append a `Booking` object to a user's profile, the user gains direct access to all the attributes encapsulated within that booking. We can calculate total revenue, check the status, or retrieve the specific `dog_id` by chaining dot notation:
`user_01.bookings[0].price`

````python
# 1. INSTANTIATE THE CORE ENTITIES
# Create the User
user_01 = User(
    name="Jacqui", 
    email="jacqui@example.com", 
    phone="555-0100"
)

# Create the Dog and link via Foreign Key
dog_01 = Dog(
    name="Waffles",
    breed="Corgi",
    weight=28.5,
    age=3.0,
    owner_id=user_01.user_id
)

# Update the User's internal state to recognize the dog
# (Maintaining our current string-based validation for this step)
user_01.dogs.append(dog_01.name)

# 2. INSTANTIATE THE TRANSACTIONAL ENTITY
# Create a formal Booking object linking both the User and the Dog
booking_01 = Booking(
    user_id=user_01.user_id,
    dog_id=dog_01.dog_id,
    service_type="daycare",
    date="2026-03-05"
)

# 3. OBJECT COMPOSITION: Appending the object to the User's history
# We append the actual object, not a string
user_01.bookings.append(booking_01)

# 4. VERIFYING THE SYSTEM STATE
print(f"--- System Status for {user_01.name} ---")
print(f"Total Registered Dogs: {len(user_01.dogs)}")
print(f"Total Bookings: {len(user_01.bookings)}\n")

print("--- Inspecting the Booking Object ---")
# Accessing the first item in the list (index 0), which is our Booking object
first_booking = user_01.bookings[0]

# Chaining dot notation to retrieve specific attributes from the nested object
print(f"Owner: {first_booking.user_id}")
print(f"Date: {first_booking.date}")
print(f"Service: {first_booking.service_type.capitalize()}")
print(f"Price: ${first_booking.price:.2f}")
print(f"Status: {first_booking.status}")
print(f"Booking Reference ID: {first_booking.booking_id}")

dog_01.get_size_category()
````

### **2.9. Advanced System Refactoring: Automating Composition**

This is a critical milestone in our system's architecture. Previously, we manually linked our entities using strings (like `owner_id`). While this mimics a database, Object-Oriented Programming (OOP) allows for a much more powerful approach: **Passing Objects as References**.

Instead of passing an ID string into the `Dog` or `Booking` classes, we will pass the entire `User` object. This grants the new object full, immediate access to the owner's attributes and methods.

#### **Key Architectural Upgrades:**

1.  **Two-Way Linking (The Dog Class)**: 
    When we instantiate a `Dog`, we will pass the actual `User` object as the owner. Inside the `Dog` constructor, the dog will automatically append *itself* (`self`) to the owner's `dogs` list. This automates the registration process and eliminates a manual step.

2.  **The Factory Method (The User Class)**: 
    We will refactor `process_booking` so it no longer just appends a string. Instead, the `User` object will act as a "Factory"—it will instantiate the `Booking` object internally and store it. 

3.  **System Logging (Print Statements)**: 
    We will add detailed `print()` statements inside the `__init__` methods of all three classes. This creates an automated audit trail, confirming object creation and relationships instantly.

Let's examine the refactored code to see these principles in action.

````python
from uuid import uuid4

# SYSTEM CONSTANTS
BOARDING_PRICE = 70.0
DAYCARE_PRICE = 50.0

class User:
    """Represents a human client. Acts as the central hub for their Dogs and Bookings."""
    
    def __init__(self, name: str, email: str, phone: str | None = None):
        self.name = name
        self.email = email
        self.phone = phone
        self.dogs = []      # Will now store actual Dog objects
        self.bookings = []  # Will now store actual Booking objects
        self.user_id = str(uuid4())
        
        # 1. AUTOMATED LOGGING: Print upon creation
        print(f"[System] Created User Profile: {self.name} | Email: {self.email} | Phone: {self.phone}")

    def update_phone(self, new_number: str):
        self.phone = new_number
        print(f"[System Update] {self.name}'s phone changed to: {self.phone}")
        
    def process_booking(self, dog_name: str, service_type: str, date_list: list):
        """Acts as a Factory Method to generate Booking objects internally."""
        print(f"\n--- Processing Request: {service_type.capitalize()} for {dog_name} ---")
        
        # VALIDATION: Search the internal list of Dog objects
        target_dog = None
        for d in self.dogs:
            if d.name == dog_name:
                target_dog = d
                break
                
        if target_dog is None:
            print(f"❌ Error: {self.name} does not have a registered dog named '{dog_name}'.")
            return

        # COMPOSITION: Instantiating the Booking object internally
        for date in date_list:
            # We pass 'self' (the User object) and 'target_dog' (the Dog object) directly
            new_booking = Booking(owner=self, dog=target_dog, service_type=service_type, date=date)
            self.bookings.append(new_booking)


class Dog:
    """Represents a pet. Automatically registers itself to its owner."""

    def __init__(self, name: str, breed: str, weight: float, age: float, owner: User):
        self.name = name
        self.breed = breed
        self.weight = weight
        self.age = age
        self.owner = owner  # Storing the entire User object reference
        self.dog_id = str(uuid4()) 
        
        # 2. AUTOMATED REGISTRATION: The dog adds itself to the owner's list
        self.owner.dogs.append(self)
        
        print(f"[System] Created Dog Profile: {self.name} ({self.breed}) | Linked to Owner: {self.owner.name} (ID: {self.owner.user_id})")
        
    def get_size_category(self):
        if self.weight < 15:
            return "Small"
        elif 15 <= self.weight < 40:
            return "Medium"
        else:
            return "Large"


class Booking:
    """Transactional entity connecting User, Dog, and Service specifics."""
    
    def __init__(self, owner: User, dog: Dog, service_type: str, date: str):
        self.owner = owner
        self.dog = dog
        self.service_type = service_type.lower()
        self.date = date
        
        if self.service_type == "boarding":
            self.price = BOARDING_PRICE
        elif self.service_type == "daycare":
            self.price = DAYCARE_PRICE
        else:
            self.price = 0.0
            
        self.booking_id = str(uuid4())
        self.status = "Confirmed"
        
        # 3. DETAILED TRANSACTION LOGGING
        print(f" ✅ [Booking Confirmed] Ref ID: {self.booking_id}")
        print(f"    Owner: {self.owner.name} ({self.owner.phone} | {self.owner.email})")
        print(f"    Pet: {self.dog.name} | Service: {self.service_type.capitalize()} | Date: {self.date}")
        print(f"    Price: ${self.price:.2f} | Status: {self.status}")
````

````python
# --- SYSTEM TEST ---
# 1. Create the User (Triggers User __init__ print)
user_01 = User("Jacqui", "jacqui@example.com", "555-0100")

# 2. Create the Dog (Triggers Dog __init__ print AND auto-appends to User)
dog_01 = Dog("Waffles", "Corgi", 28.5, 3.0, owner=user_01)

# 3. Process Bookings (Triggers Booking __init__ prints)
user_01.process_booking("Waffles", "daycare", ["2026-03-10", "2026-03-11"])
````

## **3. Data Protection and Advanced Mechanics: Python Decorators**

### **3.1. What is a Decorator?**

As our system architecture matures, we encounter scenarios where we need to modify or strictly control the behavior of functions and methods without rewriting their internal logic. In Python, this is achieved using **Decorators**.

A decorator is a structural design pattern that allows you to "wrap" existing code with additional functionality. Conceptually, a decorator acts as a gatekeeper or a modifier that intercepts a function call, executes some preliminary or post-execution logic, and then allows the original function to run. 

Syntactically, a decorator is identified by the `@` symbol placed directly above a function, method, or class definition.



### **3.2. The Decorator Ecosystem**

Because decorators are a fundamental Python feature, the ecosystem surrounding them is vast. Theoretically, an infinite number of decorators exist because any developer can write custom decorators. However, they generally fall into three distribution categories:

| Distribution Source | Availability | Description & Examples |
| :--- | :--- | :--- |
| **Built-in Python** | Immediate (No import required) | Core language features embedded directly into Python. <br>*Examples:* `@property`, `@classmethod`, `@staticmethod`. |
| **Standard Library** | Requires `import` (No installation) | Official Python modules included with the language installation, but kept out of the default namespace to save memory. <br>*Examples:* `@dataclass` (from `dataclasses`), `@lru_cache` (from `functools`). |
| **Third-Party Libraries** | Requires `pip install` & `import` | Created by the open-source community for specialized domains (Data Science, Web Engineering, AI). <br>*Examples:* `@app.route` (Flask), `@tool` (LangChain/AI). |

### **3.3. Categorizing Decorators by Use Case**

To understand their utility, let's examine specific decorators ranging from foundational object-oriented controls to advanced system engineering tools.

#### **A. Core Object-Oriented Controls (Relevant to Our Project)**
These built-in decorators are essential for **Encapsulation**—protecting the internal state of our objects from invalid data entry.

* **`@property` (The Getter):** Transforms a method so that it can be accessed like a standard attribute (e.g., `user.email` instead of `user.get_email()`). It provides read-only access to an internal variable.

* **`@<attribute>.setter` (The Gatekeeper):** Pairs with `@property` to intercept any attempt to modify an attribute (e.g., `user.email = "new@email.com"`). It allows us to inject validation logic (like checking for an "@" symbol) *before* the data is saved to memory.

* **`@dataclass` (Standard Library):** Placed above a class definition, this decorator automatically generates boilerplate code, including the `__init__` constructor, making it highly efficient for classes primarily used to store data.



#### **B. General Utility & Performance**
These decorators optimize or monitor how functions execute.

* **`@timer` (Custom):** A common custom decorator developers write to measure exactly how many milliseconds a function takes to execute, useful for identifying performance bottlenecks.

* **`@lru_cache` (Standard Library):** Stands for "Least Recently Used Cache." If a function performs a heavy mathematical calculation, this decorator forces Python to remember the result. If the function is called again with the exact same inputs, it returns the saved answer instantly instead of recalculating it.

#### **C. Advanced Engineering: Web Servers & AI Systems**
In professional production environments, decorators are the primary mechanism for routing data and exposing logic to external systems.

* **Web Routing (`@app.get("/users")`):** Used in web frameworks like FastAPI or Flask. It instructs the web server: *"When a user's browser requests the '/users' URL, execute the Python function directly below this decorator to fetch the data."*

* **Agentic AI Tools (`@tool`):** Used in AI orchestration frameworks like LangChain. By placing `@tool` above a standard Python function (e.g., our `process_booking` method), it exposes that exact function to a Large Language Model (LLM). This allows the AI agent to independently read the function's parameters and execute it to interact with our Daycare system autonomously.

---

**Next Step:** In our current architecture, a user's email or phone number can be overwritten with invalid data types (e.g., `user_01.email = 12345`). 

To demonstrate the power of Encapsulation, we will refactor the `User` class. Currently, a developer could accidentally set `user.email = "invalid-data"`, corrupting our database.

By using the `@property` and `@setter` decorators, we create a "Protected Attribute" (conventionally starting with an underscore, like `_email`). The user interacts with `email` as if it were a normal variable, but Python routes that interaction through our validation logic.

### **3.3. Refactoring for Data Integrity**

````python
from uuid import uuid4

class User:
    def __init__(self, name: str, email: str, phone: str | None = None):
        self.name = name
        # We trigger the @email.setter immediately during initialization
        self.email = email 
        self.phone = phone
        self.dogs = []
        self.bookings = []
        self.user_id = str(uuid4())
        
        print(f"[System] Created User Profile: {self.name}")

    # 1. THE GETTER (@property)
    # This allows us to access 'user.email' without parentheses.
    @property
    def email(self):
        """Read-only access to the internal _email attribute."""
        return self._email

    # 2. THE SETTER (@attribute.setter)
    # This acts as the 'Gatekeeper'. It runs every time someone tries to 
    # assign a value to 'self.email'.
    @email.setter
    def email(self, value: str):
        print(f"--- [Validation Check] Attempting to set email to: {value} ---")
        
        # BASIC VALIDATION LOGIC
        if not isinstance(value, str):
            print("❌ Validation Error: Email must be a string.")
            return
            
        if "@" not in value or "." not in value:
            print(f"❌ Validation Error: '{value}' is not a valid email format.")
            return
            
        # If all checks pass, we save it to a 'protected' internal variable
        # The underscore (_) is a convention meaning 'Internal Use Only'
        self._email = value
        print(f"✅ Email successfully validated and saved.")
````

````python

# --- TESTING THE GATEKEEPER ---

# A. Successful Initialization
user_04 = User("Alex", "alex@example.com")

# B. Attempting to set an invalid email (Missing '@')
print("\n[Test 1] Updating email to an invalid string:")
user_04.email = "alex_at_website.com" 

# C. Attempting to set an invalid data type (Integer)
print("\n[Test 2] Updating email to an integer:")
user_04.email = 123456

# D. Verifying the state remains protected
print(f"\n[Final State] {user_04.name}'s email remains: {user_04.email}")
````

### **3.4. Streamlining Code with `@dataclass`**

As you have seen, writing a robust class requires a significant amount of "boilerplate" code — repetitive tasks like setting up the `__init__` method, mapping parameters to attributes (`self.x = x`), and defining how the object should look when printed.

To streamline this, Python provides the **`@dataclass`** decorator in the Standard Library. It is specifically designed for classes whose primary purpose is data storage and management.

#### **Key Features Introduced by `@dataclass`**

1.  **Automatic Initialization**: It reads your type hints (e.g., `name: str`) and automatically generates the `__init__` method behind the scenes.

2.  **Automatic Representation**: Normally, printing a standard object outputs a cryptic memory address (e.g., `<__main__.User object at 0x104b...>`). A dataclass automatically generates a readable string: `User(name='Jacqui', email='...')`.

3.  **Handling Mutable Defaults (`default_factory`)**: In standard Python, assigning an empty list as a default parameter can cause severe bugs where multiple objects share the exact same list in memory. `@dataclass` solves this using `field(default_factory=list)`, which guarantees every new object gets its own, independent list.

Let's look at our core system entities—`Service`, `User`, and `Dog`—refactored using this powerful decorator.

````python
from dataclasses import dataclass, field
from typing import List, Optional
from uuid import uuid4

# ---------------------------------------------------------
# 1. THE SERVICE CLASS (Pure Data Container)
# ---------------------------------------------------------
@dataclass
class Service:
    """A lightweight class to manage different daycare services."""
    name: str
    base_price: float
    description: str
    
    # We use a lambda function within default_factory to execute uuid4() for each new instance
    service_id: str = field(default_factory=lambda: str(uuid4()))

# ---------------------------------------------------------
# 2. THE USER CLASS
# ---------------------------------------------------------
@dataclass
class User:
    # 1. Required Parameters
    name: str
    email: str
    
    # 2. Optional Parameters
    # 'Optional[str]' means this can be a string OR None
    phone: Optional[str] = None
    
    # 3. Dynamic Defaults (Preventing shared state bugs)
    dogs: List[str] = field(default_factory=list)
    bookings: List[str] = field(default_factory=list)
    user_id: str = field(default_factory=lambda: str(uuid4()))

    # 4. Custom Methods work exactly the same as before
    def process_booking(self, dog_name: str, service_type: str, date_list: List[str]):
        if dog_name not in self.dogs:
            print(f"❌ Error: {self.name} does not own '{dog_name}'")
            return
            
        for date in date_list:
            self.bookings.append(f"{date} | {dog_name} | {service_type}")
            print(f" ✅ Confirmed for {date}")

# ---------------------------------------------------------
# 3. THE DOG CLASS
# ---------------------------------------------------------
@dataclass
class Dog:
    name: str
    breed: str
    weight: float
    age: float
    owner_id: str
    dog_id: str = field(default_factory=lambda: str(uuid4()))

    def get_size_category(self):
        if self.weight < 15: 
            return "Small"
        elif 15 <= self.weight < 40: 
            return "Medium"
        else: 
            return "Large"

# ---------------------------------------------------------
# SYSTEM EXECUTION & VERIFICATION
# ---------------------------------------------------------

# Instantiate a User (Notice how clean the instantiation is)
user_05 = User(
    name="David", 
    email="david@example.com", 
    dogs=["Rex"]
)

# Automated Representation: Let's see how @dataclass prints the object
print("--- Automated Object Representation ---")
print(user_05)

# Verify Method Functionality
print("\n--- Testing Encapsulated Logic ---")
user_05.process_booking("Rex", "boarding", ["2026-08-15"])
````
